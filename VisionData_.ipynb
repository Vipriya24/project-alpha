## **Student's Vision Dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv('/content/sstudent_engagement_dataset.csv')
dataset
```

```python
dataset = dataset.drop(columns=["Learning_Content_Type", "Difficulty_Level"])
print(dataset)
```

```python
studentinfo=[i for i in dataset if 'Unnammed' in i]
dataset.drop(studentinfo,axis=1,inplace=True)
dataset
```

### **Check the Max and Min values in the dataset**

```python
max_value = dataset.max()
min_value = dataset.min()
print(max_value)
print(min_value)
```

```python
max_value = dataset.max()
max_value.plot()
```

```python
min_value = dataset.min()
min_value.plot()
```

### **Normalizing the Dataset**

```python
dataset.where(dataset<=2,np.nan,inplace=True)
dataset.max().plot()
```

```python
dataset.where(dataset>0.5,np.nan,inplace=True)
dataset.min().plot()
```

```python
dataset
```

### **Removing NAN values**

```python
threshold=0.8
NoNAN = dataset.notna().mean()
data=dataset.loc[:,NoNAN >= (1-threshold)]
data
```

```python
df=data.ffill(axis=1)
df=df.bfill(axis=1)
df
```

### **Task 1: Binarize the data**

```python
binary_df=(df>=1).astype(int)
binary_df
```
