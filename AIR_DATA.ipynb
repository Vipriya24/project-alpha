### Importing 2nd Data having NAN values.

```python
AIR_DATA_2=pd.read_csv('/AIR_DATA_2.csv')
AIR_DATA_2
```

```python
#To remove unnamed columns
AIR_DATA_2=AIR_DATA_2.loc[:,~AIR_DATA_2.columns.str.contains('^Unnamed')]
AIR_DATA_2
```

```python
#To remove Timestamp column which is not required for imputation
AIR_DATA_2=AIR_DATA_2.drop(columns=('Timestamp'))
AIR_DATA_2
```

**MEAN** Imputation Method:

```python
#calculate the mean of the 2nd data
for column in AIR_DATA_2.columns:
    AIR_DATA_2[column] = AIR_DATA_2[column].replace('NAN', np.nan)
    AIR_DATA_2[column] = pd.to_numeric(AIR_DATA_2[column], errors='coerce')
mean_imputed = AIR_DATA_2.fillna(AIR_DATA_2.mean())
mean_imputed.to_csv('Mean_imputed.csv', index=False)
mean_imputed
```

**MEDIAN** Imputation Method:

```python
#calculate the median of the 2nd data
def median_imputation(AIR_DATA_2):
    return AIR_DATA_2.fillna(AIR_DATA_2.median())
median_imputed = median_imputation(AIR_DATA_2)
median_imputed
```

**MODE** Imputation Method:

```python
#calculate the mode of the 2nd data
def mode_imputation(AIR_DATA_2):
    return AIR_DATA_2.fillna(AIR_DATA_2.mode().iloc[0])
mode_imputed = mode_imputation(AIR_DATA_2)
mode_imputed
```

**FORWARD FILL** Imputation Method:

```python
#fill the missing values which are replaced with the previous row's values using 'ffill'
def forward_fill_imputation(AIR_DATA_2):
    return AIR_DATA_2.fillna(method='ffill')
forward_filled = forward_fill_imputation(AIR_DATA_2)
forward_filled
```

**BACKWARD FILL** Imputation Method:

```python
#replace the NAN values with the next valid value using 'bfill'
def backward_fill_imputation(AIR_DATA_2):
    return AIR_DATA_2.fillna(method='bfill')
backward_filled = backward_fill_imputation(AIR_DATA_2)
backward_filled
```

**LINEAR** Imputation Method:

```python
#estimate the missing values using linear trends
interpolated = AIR_DATA_2.interpolate(method='linear')
interpolated
```

**CONSTANT** Imputation Method:

```python
#replace missing values with a fixed number
constant_imputed = AIR_DATA_2.fillna(0)
constant_imputed
```

**INTER-QUARTILE RANGE** Imputation Method:

```python
#replace missing values with Q1 (25th percentile) or Q3 (75th percentile)
Q1 = AIR_DATA_2.quantile(0.25)
Q3 = AIR_DATA_2.quantile(0.75)
iqr_imputed = AIR_DATA_2.fillna(Q1)
iqr_imputed
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

# Load datasets
data_original = pd.read_csv("/AIR_DATA_1.csv")
data_missing = pd.read_csv("/AIR_DATA_2.csv")

# Convert timestamp column to datetime (assuming first column is timestamp)
timestamp_col = data_missing.columns[0]  # Adjust this if needed
data_missing[timestamp_col] = pd.to_datetime(data_missing[timestamp_col], errors="coerce")

# Convert all placeholders ("NAN", "null", etc.) to actual NaN
data_missing.replace(["NAN", "null", ""], np.nan, inplace=True)

# Convert numeric columns (ignore timestamp)
for col in data_missing.columns:
    if col != timestamp_col:  # Skip timestamp column
        data_missing[col] = pd.to_numeric(data_missing[col], errors='coerce')

# Drop fully empty columns
data_missing.dropna(axis=1, how='all', inplace=True)

# Identify numeric columns (excluding the timestamp)
numeric_cols = data_missing.select_dtypes(include=[np.number]).columns.tolist()

if not numeric_cols:
    print("⚠️ No numeric columns found. Check dataset contents:")
    print(data_missing.dtypes)
    raise ValueError("❌ No numeric columns found. Ensure dataset contains numerical values.")

# Define imputation methods
imputation_methods = {
    "Mean": lambda df: df.fillna(df.mean()),
    "Median": lambda df: df.fillna(df.median()),
    "Mode": lambda df: df.apply(lambda col: col.fillna(col.mode().iloc[0]) if not col.mode().empty else col),
    "ForwardFill": lambda df: df.ffill(),
    "BackwardFill": lambda df: df.bfill(),
    "Linear Interpolation": lambda df: df.interpolate(method='linear'),
    "Constant": lambda df: df.fillna(0),  # Replace NaNs with a constant (e.g., 0)
    "Interquartile": lambda df: df.apply(lambda col: col.fillna(col.median() - 1.5 * (col.quantile(0.75) - col.quantile(0.25))))
}

rmse_results = {}

# Apply imputation methods
for method, impute_func in imputation_methods.items():
    imputed_data = data_missing.copy()

    # Apply imputation only to numeric columns
    imputed_data[numeric_cols] = impute_func(data_missing[numeric_cols])

    # 🚨 Check if NaNs still exist after imputation
    if imputed_data[numeric_cols].isna().sum().sum() > 0:
        print(f"⚠️ Warning: NaN values remain after {method} imputation")
        continue

    rmse_values = []

    for col in numeric_cols:
        missing_indices = data_missing[col].isna()
        if not missing_indices.any():
            continue  # Skip if there were no missing values initially

        original_values = data_original.loc[missing_indices, col]
        imputed_values = imputed_data.loc[missing_indices, col]

        # Ensure values are numeric & drop NaNs
        original_values = pd.to_numeric(original_values, errors='coerce').dropna()
        imputed_values = pd.to_numeric(imputed_values, errors='coerce').dropna()

        if original_values.empty or imputed_values.empty:
            print(f"⚠️ Skipping {col} in {method} due to empty values after conversion")
            continue

        rmse = np.sqrt(mean_squared_error(original_values, imputed_values))
        rmse_values.append(rmse)

    if rmse_values:
        rmse_results[method] = np.mean(rmse_values)

# Convert results to DataFrame
rmse_df = pd.DataFrame(list(rmse_results.items()), columns=["Method", "RMSE"]).sort_values(by="RMSE")

# Display results
print("\n✅ RMSE for Each Imputation Method:\n", rmse_df)

if not rmse_df.empty:
    best_method = rmse_df.iloc[0]["Method"]
    print(f"\n🏆 Best Imputation Method: {best_method} (Lowest RMSE)")
else:
    print("\n⚠️ No valid RMSE values. Check your data imputation process.")
```

Comparison for AIR_DATA_3 Sheet

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error

# Load datasets
file_original = "/AIR_DATA_1.csv"
file_third = "/AIR_DATA_3.csv"

original_data = pd.read_csv(file_original)
third_data = pd.read_csv(file_third)

# Drop unnamed columns (extra empty columns from CSV)
original_data = original_data.loc[:, ~original_data.columns.str.contains('Unnamed', na=False)]
third_data = third_data.loc[:, ~third_data.columns.str.contains('Unnamed', na=False)]

# Ensure matching columns
common_columns = original_data.columns.intersection(third_data.columns)
original_data = original_data[common_columns]
third_data = third_data[common_columns]

# Drop timestamp-related columns
timestamp_cols = [col for col in original_data.columns if "date" in col.lower() or "time" in col.lower()]
original_data_cleaned = original_data.drop(columns=timestamp_cols, errors='ignore')
third_data_cleaned = third_data.drop(columns=timestamp_cols, errors='ignore')

# Convert remaining columns to numeric
original_data_cleaned = original_data_cleaned.apply(pd.to_numeric, errors='coerce')
third_data_cleaned = third_data_cleaned.apply(pd.to_numeric, errors='coerce')

# Align datasets to the same number of rows
min_rows = min(len(original_data_cleaned), len(third_data_cleaned))
original_data_aligned = original_data_cleaned.iloc[:min_rows].reset_index(drop=True)
third_data_aligned = third_data_cleaned.iloc[:min_rows].reset_index(drop=True)

# Introduce artificial missing values (10% of the dataset)
np.random.seed(42)
third_data_with_missing = third_data_aligned.copy()
mask = np.random.rand(*third_data_with_missing.shape) < 0.1  # 10% missing values
third_data_with_missing[mask] = np.nan

# Identify missing value columns
missing_cols = third_data_with_missing.columns[third_data_with_missing.isnull().any()]

# Define imputation function
def impute_and_evaluate(method, df_original, df_missing):
    df_imputed = df_missing.copy()

    if method == 'mean':
        df_imputed = df_imputed.fillna(df_imputed.mean())
    elif method == 'median':
        df_imputed = df_imputed.fillna(df_imputed.median())
    elif method == 'mode':
        df_imputed = df_imputed.apply(lambda col: col.fillna(col.mode().iloc[0]) if col.isnull().any() else col)
    elif method == 'forward_fill':
        df_imputed = df_imputed.ffill()
    elif method == 'backward_fill':
        df_imputed = df_imputed.bfill()
    elif method == 'linear':
        df_imputed = df_imputed.interpolate(method='linear')
    elif method == 'constant':
        df_imputed = df_imputed.fillna(0)  # Replace NaNs with 0
    elif method == 'interquartile':
        for col in missing_cols:
            Q1 = df_imputed[col].quantile(0.25)
            Q3 = df_imputed[col].quantile(0.75)
            df_imputed[col] = df_imputed[col].fillna((Q1 + Q3) / 2)

    # Ensure no NaN remains before evaluating
    if df_imputed.isnull().sum().sum() > 0:
        raise ValueError(f"NaN values remain after {method} imputation!")

    # Evaluate using Mean Absolute Error (MAE)
    mae = mean_absolute_error(df_original[missing_cols], df_imputed[missing_cols])
    return mae

# Test all imputation methods
methods = ['mean', 'median', 'mode', 'forward_fill', 'backward_fill', 'linear', 'constant', 'interquartile']
results = {}

for method in methods:
    try:
        results[method] = impute_and_evaluate(method, original_data_aligned, third_data_with_missing)
    except Exception as e:
        print(f"Error in {method} imputation: {e}")
        results[method] = float("inf")  # Assign a high error value if method fails

# Determine the best method
best_method = min(results, key=results.get)
print("Best imputation method:", best_method)
print("Imputation method errors:", results)
```
